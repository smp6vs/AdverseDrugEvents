#### Confusion matrix set up

```python
## classifiers we are using for classification
CLASSIFIER_NAMES = ["SVM", "Decision Tree", "Random Forest", "Neural Networks"]
CLASSIFIERS = [LinearSVC(C=0.1, loss='hinge', penalty='l2', random_state=RANDOM_STATE),  ## using best setting from part 1
               DecisionTreeClassifier(max_depth=3,random_state=RANDOM_STATE),  ## keeping the depth 3 for visualization purposes 
               RandomForestClassifier(max_depth=10, n_estimators=100,random_state=RANDOM_STATE),
               LogisticRegression(C=0.01, random_state=RANDOM_STATE)]

def displayConfusionMatrix(confusionMatrix, title):
    plt.rcParams['figure.figsize'] = [16, 16] ## plot size
    confusionMatrix = np.transpose(confusionMatrix)
    precisionLow = round((confusionMatrix[0][0] / (confusionMatrix[0][0] + confusionMatrix[0][1]))*100, 1)
    precisionHigh = round((confusionMatrix[1][1] / (confusionMatrix[1][0] + confusionMatrix[1][1]))*100, 1)
    recallLow = round((confusionMatrix[0][0] / (confusionMatrix[0][0] + confusionMatrix[1][0]))*100, 1)
    recallHigh = round((confusionMatrix[1][1] / (confusionMatrix[0][1] + confusionMatrix[1][1]))*100, 1)

    true_pos = np.diag(confusionMatrix)
    plt.imshow(confusionMatrix, interpolation='nearest',cmap=plt.cm.Blues,vmin=0, vmax=100)
    xticks = np.array([0,1])
    plt.gca().set_xticks(xticks)
    plt.gca().set_yticks(xticks)
    plt.gca().set_xticklabels(["High " + str(LABEL_VAL) + "\n Recall=" + str(recallLow), "Low "  + str(LABEL_VAL) + "\n Recall=" + str(recallHigh)], fontsize=PLOT_FONT_SIZE-2)
    plt.gca().set_yticklabels(["High " + str(LABEL_VAL) + "\n Precision=" + str(precisionLow), "Low " + str(LABEL_VAL) + "\n Precision=" + str(precisionHigh)], fontsize=PLOT_FONT_SIZE-2)
    plt.ylabel("Predicted Class", fontsize=PLOT_FONT_SIZE-1)
    plt.xlabel("Actual Class", fontsize=PLOT_FONT_SIZE-1)
    plt.title(title, fontsize=PLOT_FONT_SIZE)
        
    ## add text in heatmap boxes
    for i in range(len(xticks)):
        for j in range(len(xticks)):
            text = plt.text(j, i, confusionMatrix[i][j], ha="center", va="center", color="white", size=15) ### size here is the size of text inside a single box in the heatmap

def trainClassifier(trainFeatures, trainLabels, testFeatures, testLabels,classificationReport = False, classificationSetting = ""):
    """Used to train a number of classifiers and return results"""
    global rocCurveData
    
    ## empty lists for storing results
    testAccuracy, testPrecision, testRecall, testAUC = [], [], [], []
    
    ## list to store importance value for each feature
    featureImportance = []
    
    ## variable to store the trained classifier for visualization purposes
    classifierFile = None
    
    ## iterate through classifiers defined above
    classifierIterationCount = 0
    for name, classifier in zip(CLASSIFIER_NAMES, CLASSIFIERS):
            
        ## training the classifier
        _ = classifier.fit(trainFeatures, trainLabels)

        ## obtaining test accuracy
        score = classifier.score(testFeatures, testLabels) * 100
            
        ## decision tree is a good example of an "interpretable model". we will see a visualization of the model later
        if name == "Decision Tree":
            classifierFile = copy.deepcopy(classifier)
        
        ## more test metrics
        predictions = classifier.predict(testFeatures)
        if name == "SVM":
            predictionsProbabilities = classifier.decision_function(testFeatures)  ## svm confidence retrieval function is different than others
        else:
            predictionsProbabilities = classifier.predict_proba(testFeatures)
            predictionsProbabilities = [item[1] for item in predictionsProbabilities]
            
        precision = precision_score(testLabels, predictions, average=None)
        recall = recall_score(testLabels, predictions, average=None)
        aucScore = roc_auc_score(testLabels, predictionsProbabilities) * 100

        ## we only save importance values from random forest
        if name == "Random Forest":
            featureImportance = classifier.feature_importances_
        
        ## we only save prediction confidences for ROC analysis for SVM, here
        if name == "SVM": 
            predictionsConfidence = classifier.decision_function(testFeatures)
            fpr, tpr, threshold = roc_curve(testLabels, predictionsConfidence)
            settings = [item[0] for item in rocCurveData]
            if classificationSetting not in settings:
                rocCurveData.append([classificationSetting, fpr, tpr])
        
        ## printing confusion matrix
        if classificationReport == True:
            confusionMatrix = confusion_matrix(testLabels, predictions)
            plt.subplot(int(str(22) + str(classifierIterationCount+1)))
            displayConfusionMatrix(confusionMatrix, name)
            classifierIterationCount = classifierIterationCount + 1

        testAccuracy.append(score)
        testPrecision.append(precision)
        testRecall.append(recall)
        testAUC.append(aucScore)
        
    ## show confusion matrix if this variable is true
    if classificationReport == True:
        plt.subplots_adjust(hspace = 0.5)
        plt.show()
        
    ## return results
    return testAccuracy, testPrecision, testRecall, testAUC, featureImportance, classifierFile
```

Convert the text chunks to all lower case. Remove the stopwords (such as 'a', 'the', and 'to'). Keep each word's lemma, e.g., the verbs gone, going, and went have the lemma go. The lemmatizer needs to know the word's part of speech (pos). Use the regular expressions package re to remove numbers, punctuation, and whitespace.

```python
%%time
wordnet_lemmatizer = WordNetLemmatizer()
stopset = list(set(stopwords.words('english')))
clean_reviews_text = []
for review in inputDataTrain['Text']:  # Loop through the tokens (the words or symbols) in each review. 
    try:    
        cleaned_review = re.sub("[^a-zA-Z]"," ", review)  # Remove numbers and punctuation.
        cleaned_review = cleaned_review.lower()  # Convert the text to lower case.
        cleaned_review = ' '.join([word for word in cleaned_review.split() if word not in stopset])  # Keep only words that are not stopwords.
        cleaned_review = ' '.join([wordnet_lemmatizer.lemmatize(word, pos='n') for word in cleaned_review.split()])  # Keep each noun's lemma.
        cleaned_review = ' '.join([wordnet_lemmatizer.lemmatize(word, pos='v') for word in cleaned_review.split()])  # Keep each verb's lemma.
        cleaned_review = re.sub(r"(http\S+)"," ", cleaned_review)  # Remove http links.
        cleaned_review = ' '.join(cleaned_review.split())  # Remove white space.
    except TypeError:
        pass
    clean_reviews_text.append(cleaned_review)
```

### Character Convolutional Neural Network

```python
MAX_CHARACTERS_TO_KEEP = 500
def convertCharDataIntoArrays(inputText, characters, charsDict, MAX_CHARACTERS_TO_KEEP):
    characterFeatures = []
    for row in inputText:
        characterFeature = np.zeros((MAX_CHARACTERS_TO_KEEP, len(characters)+1))
        for c in range(0, min(len(row), MAX_CHARACTERS_TO_KEEP)):
            character = row[c]
            ## explicitly converting each character into a character embedding i.e just a one hot encoded representation
            charPresence = np.zeros(len(characters)+1)
            if character in charsDict:
                charPresence[charsDict[character]] = 1
                characterFeature[c] = charPresence
        characterFeatures.append(characterFeature)
        
    return np.array(characterFeatures)

characters = " ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789-,'.!?();:-_=+@#$%^&*/\""
charsDict = {}
numsDict = {}

## assign an id to each character
for c in range(0,len(characters)):
    charsDict[characters[c]] = c+1
    numsDict[c+1] = characters[c]
    

## create feature vectors for each row of text (Train)
characterFeaturesTrain = convertCharDataIntoArrays(trainText, characters, charsDict, MAX_CHARACTERS_TO_KEEP) 
    
## create feature vectors for each row of text (Test)
characterFeaturesTest = convertCharDataIntoArrays(testText, characters, charsDict, MAX_CHARACTERS_TO_KEEP) 
    
## create feature vectors for each row of text (Validation)
characterFeaturesValidation = convertCharDataIntoArrays(validationText, characters, charsDict, MAX_CHARACTERS_TO_KEEP) 

print('Shape of array for character features', characterFeaturesTrain.shape)
```

```python
## this is exactly similar to word convolutional neural network except we are using characters as inputs
## create basic cnn model
charInput = Input(shape=characterFeaturesTrain.shape[1:], dtype='float32', name='charInput')
structuredDataInput = Input(shape=trainCombined.shape[1:], dtype='float32', name='structuredDataInput')

## neural network for structured data
structuredNN = layers.Dense(64, input_shape=trainCombined.shape[1:])(structuredDataInput)
structuredNN = layers.Dropout(0.5)(structuredNN)

## char convolutional neural network
charCNN = layers.Conv1D(64, 5, activation='relu',input_shape=characterFeaturesTrain.shape[1:])(charInput)
charCNN = layers.Dropout(0.2)(charCNN)
charCNN = layers.GlobalMaxPooling1D()(charCNN)
charCNN = layers.Dropout(0.2)(charCNN)

## concatenate both networks together
if WITH_STRUCTURED_DATA:
    concatenation = keras.layers.concatenate([charCNN, structuredNN])
    concatenation = layers.Dropout(0.2)(concatenation)
    denseLayer = layers.Dense(64)(concatenation)
else:  
    ## no concatenation
    denseLayer = layers.Dense(64)(charCNN)

output = layers.Dense(1, activation='sigmoid')(denseLayer)

model = Model(inputs=[charInput,structuredDataInput], outputs=output)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("********* Training the model ******************")
EPOCHS = 10
bestTestAccuracyForCharModel = 0.0
bestValidationAccuracy = 0.0
losses = []
accuracies = []
trainingAccuracies = []
aucScoreChar, precisionLowChar, precisionHighChar, recallLowChar, recallHighChar, fpr, tpr = None, None, None, None, None, None, None ##initializing best test metrics
for epoch in range(0, EPOCHS):
    model.fit([characterFeaturesTrain, trainCombined], trainLabels,
                        epochs=1,
                        verbose=2,
                        batch_size=10)
    
    ## test on validation set
    loss, accuracy = model.evaluate([characterFeaturesValidation, validationCombined], validationLabels, verbose=False)
    losses.append(loss)
    accuracies.append(accuracy)
    if accuracy > bestValidationAccuracy:
        bestValidationAccuracy = accuracy
        _, bestTestAccuracyForCharModel = model.evaluate([characterFeaturesTest, testCombined], testLabels, verbose=False)
        
        ## get metrics
        predictionsProbs = model.predict([characterFeaturesTest, testCombined])
        predictions = [1 if value >= 0.5 else 0 for value in predictionsProbs]
        aucScoreChar, precisionLowChar, precisionHighChar, recallLowChar, recallHighChar, fpr, tpr = getMetrics(testLabels, predictions, predictionsProbs)

bestTestAccuracyForCharModel = bestTestAccuracyForCharModel * 100
bestValidationAccuracy = bestValidationAccuracy * 100
print("Best Validation Accuracy for Char Model: %.1f " % (bestValidationAccuracy)) ### .1f means just use one decimal point in float 
print("Test Accuracy for Char Model: %.1f " % (bestTestAccuracyForCharModel))
print("Test Precision Low %.1f, Precision High %.1f, Recall Low %.1f, Recall High %.1f, AUC Score %.1f" % 
                          (precisionLowChar,precisionHighChar,recallLowChar,recallHighChar,aucScoreChar))

alreadyPresentItems = [item[0] for item in rocCurveList] ## when we run the same cell multiple times, we don't have to add fpr and tpr again
if "CharCNN" not in alreadyPresentItems:
    rocCurveList.append(["CharCNN",fpr,tpr])


plotResults(losses, accuracies)
plt.show()
```

# Set up the scoring rule

Write a function to calculate the average Brier score. The Brier score is 1 minus the mean squared error.

The Brier score is a proper score function that measures the accuracy of probabilistic predictions. It is applicable to tasks in which predictions must assign probabilities to a set of mutually exclusive discrete outcomes.

```python
np.mean(y_train)
```

```python
def brier_score(predictions, realizations):
    this_brier_score = 1 - brier_score_loss(realizations, predictions)
    return this_brier_score
```

Write a function to calculate the skill score, your predictions' percentage toward perfect above and beyond the naive forecast. In other words, how much of the gap toward perfection did your model close?

```python
def skill_score(predictions, realizations):
    naive = np.repeat(np.mean(y_train), len(realizations))
    this_skill_score = (brier_score(predictions, realizations) - brier_score(naive, realizations)) / (1 - brier_score(naive, realizations))
    return this_skill_score
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import * ## importing all metrics
import numpy as np    ## vector and matrix computations
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import * ## SVM Classifiers
```

```python
PLOT_FONT_SIZE = 10    #font size for axis of plots
rocCurveDataAll = [] # this array stores all the ROC input data across classifiers and settings
_fpr, _tpr, _threshold = None, None, None

#define helper function for storing ROC values across settings

def appendROCValues(CLASSIFIER_NAME, MODEL_NAME, testLabels, predictionsConfidence): 
    #"""ROC value append"""
    if CLASSIFIER_NAME == "SVM":
        print("SVM ROC") #since SVM array is 1D, need this If statement here
    else:
        predictionsConfidence = [item[1] for item in predictionsConfidence] 
    
    _fpr, _tpr, _threshold = roc_curve(testLabels, predictionsConfidence)
    _aucScore = round(roc_auc_score(testLabels, predictionsConfidence) * 100,1)

    rocCurveDataAll.append([CLASSIFIER_NAME, MODEL_NAME, _fpr, _tpr, _aucScore])
```

```python
#SVM
# note that SVM is different from RF, XGB, and DT in that you need to call decision_function instead of predict_proba
# don't worry about the models/settings being used here. Point is to show Models A, B, C, D, E, etc...
# for your use cases, a grid search or random search across parameter combinations can be used
# this code trains the models, runs on test data, and sends predictions to ROC function (appendROCValues)

#define SVM parameter settings, if any (none right now...running 3x2 grid search)

CLASSIFIER_NAME = "SVM"
LOSS_FUNCTION = 'squared_hinge' # note, some parameter combinations such as l1 and hinge are not supported
PENALTY_LOSS = 'l2'

try:
    # Setting A
    clfA = LinearSVC(C=0.1, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfA.fit(X_train, y_train)
    predictionProb = clfA.decision_function(X_test)
    MODEL_NAME = "Setting A"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions
                    
    # Setting B
    clfB = LinearSVC(C=0.5, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfB.fit(X_train, y_train)
    predictionProb = clfB.decision_function(X_test)
    MODEL_NAME = "Setting B"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions

    # Setting C
    clfC = LinearSVC(C=1.0, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfC.fit(X_train, y_train)
    predictionProb = clfC.decision_function(X_test)
    MODEL_NAME = "Setting C"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions

    # Setting D
    clfD = LinearSVC(C=1.5, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfD.fit(X_train, y_train)
    predictionProb = clfD.decision_function(X_test)
    MODEL_NAME = "Setting D"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions
                    
    # Setting E
    clfE = LinearSVC(C=2.0, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfE.fit(X_train, y_train)
    predictionProb = clfE.decision_function(X_test)
    MODEL_NAME = "Setting E"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions

    # Setting F
    clfF = LinearSVC(C=2.5, loss=LOSS_FUNCTION, penalty=PENALTY_LOSS, max_iter=1000)
    clfF.fit(X_train, y_train)
    predictionProb = clfF.decision_function(X_test)
    MODEL_NAME = "Setting F"
    appendROCValues(CLASSIFIER_NAME, MODEL_NAME, y_test, predictionProb) #appends to ROC array: setting, testLabels, predictions

except Exception as e:
    ## some parameter combinations don't work, therefore, we ignore them
    print("Had to catch exception. Some parameter combinations don't work for SVM")
```

```python
class SentiWordNetCorpusReader:
    def __init__(self, SentiWordNet_3):
        """
        Argument:
        filename -- the name of the text file containing the
                    SentiWordNet database
        """        
        self.filename = filename
        self.db = {}
        self.parse_src_file()

    def parse_src_file(self):
        lines = codecs.open(self.filename, "r", "utf8").read().splitlines()
        lines = filter((lambda x : not re.search(r"^\s*#", x)), lines)
        for i, line in enumerate(lines):
            fields = re.split(r"\t+", line)
            fields = map(str.strip, fields)
            try:            
                pos, offset, pos_score, neg_score, synset_terms, gloss = fields
            except:
                sys.stderr.write("Line %s formatted incorrectly: %s\n" % (i, line))
            if pos and offset:
                offset = int(offset)
                self.db[(pos, offset)] = (float(pos_score), float(neg_score))

    def senti_synset(self, *vals):        
        if tuple(vals) in self.db:
            pos_score, neg_score = self.db[tuple(vals)]
            pos, offset = vals
            synset = wn._synset_from_pos_and_offset(pos, offset)
            return SentiSynset(pos_score, neg_score, synset)
        else:
            synset = wn.synset(vals[0])
            pos = synset.pos
            offset = synset.offset
            if (pos, offset) in self.db:
                pos_score, neg_score = self.db[(pos, offset)]
                return SentiSynset(pos_score, neg_score, synset)
            else:
                return None

    def senti_synsets(self, string, pos=None):
        sentis = []
        synset_list = wn.synsets(string, pos)
        for synset in synset_list:
            sentis.append(self.senti_synset(synset.name))
        sentis = filter(lambda x : x, sentis)
        return sentis

    def all_senti_synsets(self):
        for key, fields in self.db.items():
            pos, offset = key
            pos_score, neg_score = fields
            synset = wn._synset_from_pos_and_offset(pos, offset)
            yield SentiSynset(pos_score, neg_score, synset)
```

```python
class SentiSynset:
    def __init__(self, pos_score, neg_score, synset):
        self.pos_score = pos_score
        self.neg_score = neg_score
        self.obj_score = 1.0 - (self.pos_score + self.neg_score)
        self.synset = synset

    def __str__(self):
        """Prints just the Pos/Neg scores for now."""
        s = ""
        s += self.synset.name + "\t"
        s += "PosScore: %s\t" % self.pos_score
        s += "NegScore: %s" % self.neg_score
        return s

    def __repr__(self):
        return "Senti" + repr(self.synset)
```
